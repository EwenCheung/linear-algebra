### Basic Operation ( + , - , *, @, transpose)

```python
A = Matrix([[1, 2], 
            [3, 4]])
B = Matrix([[5, 6], 
            [7, 8]])

# Addition
print("A - 2B =")
display(A - 2* B)

# Scalar Multiplication ( for divide , only can A/n , can also use 1/n * A) 
print("2 * A =")
display(2 * A)

# Matrix Multiplication
print("A @ B =")
display(A @ B)

# Transpose
print("A^T =")
display(A.T)
```

### Solve Linear System

```python
A = Matrix([[1, 2, 3], 
            [4, 5, 5], 
            [7, 8, 9]])

b = Matrix([1, 
            2, 
            3])

A.row_join(b).rref()
display(A.row_join(b).rref()) # can view answer by rref or 
x = A.solve(rhs=b)[0] # directly use solve function
display(x)
```

```python
A = Matrix([[1, 2, 3, 4, 6], 
            [4, 5, 5, 5, 8], 
            [7, 8, 9, 8, 9]])

b = Matrix([1, 
            2, 
            3])

# Solve directly
R = A.row_join(b)
display(R.rref())
x = A.solve(rhs=b)[0]
display(x)

R.nullspace()
```
